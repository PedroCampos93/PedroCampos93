```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import QuantileTransformer

# leitura de dados
empresas = pd.read_csv ('/content/empresas.csv')
descricao = pd.read_html('https://www.contabilizei.com.br/contabilidade-online/cnae/')[0]


# tranformacao de dados
empresas['estado_empresa'] = empresas.loc[:, 'estado_empresa'].str.upper()
empresas['cnae_empresa'] = empresas['cnae_empresa'].str.replace('-', '')
empresas['cnae_empresa'] = empresas['cnae_empresa'].str.replace('.', '')


descricao = descricao.drop(columns=['Anexo', 'Fator R', 'Alíquota', 'Contabilizei'])
descricao['CNAE'] = descricao['CNAE'].astype('str')

# descricao merge
empresas_desc = empresas.merge(descricao, left_on='cnae_empresa', right_on='CNAE', how='left')


# normalizando rating
qt = QuantileTransformer(random_state=42)
var_transform = qt.fit_transform(empresas_desc[['capital_empresa', 'processos_judiciais_empresa', 'idade_empresa']])
empresas_desc[['capital_norm', 'processos_norm', 'idade_norm']] = var_transform
empresas_desc['rating'] = empresas_desc['capital_norm']*2 + empresas_desc['idade_norm']*1 + empresas_desc['processos_norm']*-1.5
empresas_desc = empresas_desc.sort_values('rating', ascending=False)

empresas_desc.to_csv('empresas_transform.csv', index=False)

```
